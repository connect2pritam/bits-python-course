```python
font1 = {'family': 'serif', 'color':'red', 'size':20 }
font2 = {'family': 'serif', 'color':'blue', 'size':10 }

xpoints = np.array([1,2,4,6])
ypoints = np.array([3,4,6,8])

plt.xlabel("This is x axis", fontdict=font2)
plt.ylabel("this is y axis",fontdict=font2)
plt.title("this is the title", fontdict=font1)
plt.plot(xpoints,ypoints,linestyle='dashed', color='r')
plt.show()
```
